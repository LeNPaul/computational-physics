```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('data.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        y.append(row[6])

for i in range(5,len(y),5):
    x.append(i)

plt.scatter(x[0:10],y[10:20])
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```
