```python
import matplotlib.pyplot as plt
import numpy as np
import csv

x = []
y = []

with open('data.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(csvfile)
    for row in plots:
        y.append(row[0])

T2 = []
        
for i in range(len(y)):
    T2.append(float(y[i]))
    x.append(i)

print(np.mean(T2))
print(np.std(T2))
    
params = {'text.usetex': True}    
    
plt.plot(x,y)
plt.xlabel('Time Step')
plt.ylabel('Energy')
plt.title('Average Total Energy Per Time Step')
plt.show()
```
